```python
def doc_per_pop(doc, pop, city):
    doc['기간'] = doc['기간'].astype(str)
    years = doc['기간'].unique()
    li = []
    for year in years:
        df_doc = doc[doc['구별'] == city]
        doc_value = df_doc[df_doc['기간'] == str(year)]["계"].values
        te = pop[pop["기간"].str.contains(str(year))]
        pop_value = te[te['구별'] == str(city)]['인구'].mean()
        result = round((doc_value[0]/pop_value) * 100, 2)
        li.append([city, result])
    return li
```

```python
car = pd.read_csv("NewSeoulCar.csv",index_col=1)
```

```python
car =car.reset_index().drop("Unnamed: 0",axis=1)
```

```python
car = car.rename(columns={"자동차 등록 합계":"합계"})
```

```python
car = car[car['구별'] != '서울시']
```

```python
car['기간'] = car['기간'].astype(str)
```

```python
cities = np.sort(car['구별'].unique())
```

```python
li = []
for city in cities:
    for year in ["2018","2019","2020"]:
        car_y = car[car['기간'].str.contains(year)]
        total = car_y[car_y['구별'] == city]['합계'].mean()
        li.append([year,city,round(total)])
li
```

```python
df_car = pd.DataFrame(li)
```

```python
df_car = df_car.rename(columns={0:"기간",1:"구별",2:"합계"})
```

```python
df_car = df_car.sort_values(by="기간").reset_index().drop("index",axis=1)
```

```python
df_car.to_csv("car.csv")
```
