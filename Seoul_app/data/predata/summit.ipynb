```python
city
```

```python
latlng = []
for i in city:
    lat = df[df['구별']==i]['위도'].mean()
    lng = df[df['구별']==i]['경도'].mean()
    latlng.append([lat,lng])
latlng
```

```python
latlng_metro = []
for i in metro_sort:
    lat = metro[metro['구별']==i]['위도'].mean()
    lng = metro[metro['구별']==i]['경도'].mean()
    latlng_metro.append([lat,lng])
latlng_metro
```

```python
latlng_metro[0]
```

```python
latlng[0]
```

```python
result = []

for ci,me  in zip(latlng,latlng_metro):
    a = (ci[0] - me[0])**2
    b = (ci[1] - me[1])**2
    c = (a+b)**0.5
    result.append(c)
result
```

```python
std = np.array(result).max()
result
```

```python
result_val = []
for i in result:
    result_val.append(round((i/std)*100,1))

result_val
```

```python
target = np.array(result_val).min()

ree = []
for i in result_val:
    ree.append(round((target/i)*10,1))
ree
```

```python
metro_dict = {}
for ci,t in zip(city, ree):
    metro_dict.update({ci:t})

metro_dict
```

```python
me = pd.DataFrame(metro_dict.items(),columns=['구별',"점수"])
```
